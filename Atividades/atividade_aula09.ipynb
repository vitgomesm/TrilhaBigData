```python
import numpy as np
import pandas as pd
import matplotlib as plt
import random
```

```python
#Criando a array de 50 n√∫meros

nums = [random.randint(1, 444) for n in range (50)]
numeros=np.array(nums) 
print(numeros)
```

```python
#Transformando dados em DataFrame
numeros_df=pd.DataFrame(numeros, columns=['Dados'])
print(numeros_df)
```

```python
#Pegando as medidas
valor_max = numeros_df['Dados'].max()
valor_min = numeros_df['Dados'].min()
amplitude = valor_max - valor_min

print(f'Maior Valor: {valor_max}')
print(f'Menor Valor: {valor_min}')
print(f'Amplitude: {amplitude}')
```

```python
#Criando o Gr√°fico
import matplotlib.pyplot as plt
fig,ax = plt.subplots(1, 2, figsize = (12, 6))

ax[0].boxplot(numeros_df['Dados'])
ax[0].set_title('Distribui√ß√£o de Dados - Boxplot')
ax[0].set_ylabel('Dados')

ax[1].text(0.1, 0.8, f"Menor Valor: {valor_min}")
ax[1].text(0.1, 0.7, f"Maior Valor: {valor_max}")
ax[1].text(0.1, 0.6, f"Amplitude: {amplitude}")
ax[1].set_axis_off()

plt.tight_layout()
plt.show()
```
