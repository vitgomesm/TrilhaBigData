```python
import numpy as np
import pandas as pd
import matplotlib as plt
```

**

```python
amostra=np.array([2500, 3200, 2700, 3000, 3700, 4500, 5000, 5200, 6000, 7000])
```

```python
amostra_df=pd.DataFrame(amostra, columns=['Valores'])
print(amostra_df)
```

```python
#Medidas de Dispersão 
valor_maximo = amostra_df['Valores'].max()
valores_minimo = amostra_df['Valores'].min()
amplitude_total = valor_maximo - valores_minimo

print(f"Maior Valor: {valores_minimo}")
print(f"Menor Valor: {valor_maximo}")
print(f"Amplitude: {amplitude_total}")
```

**Visualização do BoxPlot e Medidas:**

```python
import matplotlib.pyplot as plt
#configuração do painel
fig,ax = plt.subplots(1, 2, figsize = (12, 6))

#Boxplot
ax[0].boxplot(amostra_df['Valores'])
ax[0].set_title('Distribuição de Dados - Boxplot')
ax[0].set_ylabel('Valores')

#Exibindo Medidas
ax[1].text(0.1, 0.8, f"Menor Valor: {valores_minimo}")
ax[1].text(0.1, 0.7, f"Maior Valor: {valor_maximo}")
ax[1].text(0.1, 0.6, f"Amplitude total: {amplitude_total}")
ax[1].set_axis_off()


plt.tight_layout()
plt.show()
```
