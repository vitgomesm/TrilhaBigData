**Quartis em Código**

```python
import pandas as pd
import numpy as np
import random
```

```python
amostra=[random.randint(1,99) for _ in range (20)]
dados=np.array(amostra)
print(dados)
```

```python
#Cálculo dos Quartis:

q1=np.percentile(dados,25) # o primeiro quartil representa 25% do total da amostra
q2=np.percentile(dados,50) # mediana pode ser interpretada como o seu segundo quartil (quartil 50)
q3=np.percentile(dados,75) # o terceiro quartil representa do total da amostra

#Resumo Estátistico
media = np.mean(dados)
mediana = np.median(dados)
distancia = (media-mediana)/mediana*100

#Exibição:
print(f"Média: {media}")
print(f"Mediana: {mediana}")
print(f"Primeiro Quartil (Q1): {q1}")
print(f"Segundo Quartil (Q1): {q2}")
print(f"Terceiro Quartil (Q1): {q3}")
```
